# 九九乘法表


```python
def multiplication_table(m, n):
    print ('\n'.join(['\t'.join(['{}*{}={}'.format(i,j,i*j) for i in range(m,n+1)]) for j in range(1,10)]))
    
multiplication_table(1, 3)
print()
multiplication_table(4, 6)
print()
multiplication_table(7, 9)
```

# 金字塔

```python
def pyramid(n):
    for i in range(1,n+1):
        s=" "*(n-i)+"*"*(2*i-1)
        print(s)
pyramid(10)
```
